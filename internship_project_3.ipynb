## **Import Library**

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

## **Import Dataset**

```python
df = pd.read_csv('https://raw.githubusercontent.com/YBI-Foundation/Teaching-Data/main/Women%20Clothing%20E-Commerce%20Review.csv', encoding="ISO-8859-1")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

## **Missing Values**

remove missing values in review columns with no review text

```python
df.isna().sum()
```

```python
df[df['Review']==""]=np.NaN
```

```python
df['Review'].fillna("No Review", inplace=True)
```

```python
df.isna().sum()
```

```python
df['Review']
```

## **Define Target(y)and feature(x)**

```python
df.columns
```

```python
x=df['Review']
```

```python
y=df['Rating']
```

```python
df['Rating'].value_counts()
```

## **Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, stratify = y, random_state = 2529)
```

```python
 X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## **get features text conversion for tokens**

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer (lowercase = True, analyzer='word', ngram_range=(2, 3), stop_words ='english', max_features=5000)
```

```python
X_train=cv.fit_transform(X_train)
```

```python
cv.get_feature_names_out()
```

```python
X_train.toarray()
```

```python
X_test=cv.fit_transform(X_test)
```

```python
cv.get_feature_names_out()
```

```python
X_test.toarray()
```

## **get model train**

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model =MultinomialNB()
```

```python
model.fit(X_train,y_train)
```

## **get Model prediction**

```python
y_pred =model.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

## **get probability of each Predicted class**

```python
model.predict_proba(X_test)
```

## **Get Model Evaluation**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
